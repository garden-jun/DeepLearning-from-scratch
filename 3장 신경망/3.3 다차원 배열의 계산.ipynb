## **3.3.1 다차원 배열**
숫자를 N차원으로 나열하는 것을 통틀어 다차원 배열이라고 한다.

```python
# 1차원배열
import numpy as np
A = np.array([1,2,3,4])
print(A)            # [1 2 3 4]

print(np.ndim(A))   # 배열의 차원 수 확인 # 1

print(A.shape)      # 배열의 형상   # (4, )
print(A.shape[0])   # 4
```

```python
# 2차원 배열
B = np.array([[1,2], [3,4], [5,6]])
print(B)

print(np.ndim(B))  # 2
print(B.shape)     # (3, 2)
```

## **3.3.2 행렬의 내적(행렬 곱)**

행렬의 내적을 구하는법  
image.png  
그림에서처럼 왼쪽 행렬의 행과 오른쪽 행렬의 열을 원소별로 곱하고 그 값들을 더해서 계산합니다.

다음은 행렬곱 계산을 파이썬으로 구현 한 것입니다.

```python
A = np.array([[1,2], [3,4]])
print(A.shape)   # (2,2)

B = np.array([[5,6], [7,8]])
print(B.shape)   # (2,2)

print(np.dot(A, B))  # 행렬의 내적을 계산하는 함수
```

```python
# 2X3 행렬과 3X2 행렬의 곱
A = np.array([[1,2,3], [4,5,6]])
print(A.shape)  # (2,3)

B = np.array([[1,2], [3,4], [5,6]])
print(B.shape)  # (3,2)

print(np.dot(A, B)) # 행렬 곱 출력
```

행렬곱은 앞행렬의 열수와 뒤 행렬의 행수 가 일치해야 한다.  

    A X B = C일때
    C의 형상은 A의 행 수와 C의 열 수가 된다.
